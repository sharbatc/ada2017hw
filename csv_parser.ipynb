```python
df_masslog_formatted = pd.DataFrame(df_masslog.iloc[:,:4]);

for i in range(2,int(df_masslog.shape[1]/2)):
    new_slice = pd.DataFrame(df_masslog.iloc[:,[0,1,2*i,2*i+1]])
    new_slice.columns = ['animal_ID','exp_ID','date','mass']
    print(new_slice)
    df_masslog_formatted = df_masslog_formatted.append(new_slice, ignore_index = True)
```

## Plotting the weights

```python
grouped = df_masslog_formatted.groupby('animal_ID')

ncols = 3
nrows = int(np.ceil(grouped.ngroups/ncols))
```

```python
%matplotlib inline  
```

```python
matplotlib.rcParams.update({'font.size': 22})
fig, ax = plt.subplots(figsize = (20,10))
fig.suptitle('Mass Log')
plt.xlabel('date')
plt.ylabel('mass (gm)')
df_masslog_formatted.groupby('animal_ID')['mass'].plot(legend=True);
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.grid()
```

```python
animal_id = 'DO07'
df_masstrace = df_masslog_formatted.loc[df_masslog_formatted['animal_ID'] == animal_id]

fig, ax = plt.subplots(figsize = (20,10))
fig.suptitle(animal_id)
plt.xlabel('date')
plt.ylabel('mass (gm)')
df_masstrace.plot(y = 'mass', ax = ax)
ax.set_axisbelow(True)
ax.grid()
```
